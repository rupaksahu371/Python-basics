```python
#dictionary in python

diction ={'name':'rupak', 'age':22, 'Address':'Durg (C.G)'}
print(type(diction))

print(diction['age']) #getting key or values with different function
print(diction.get('name'))
print(diction.keys())
print(diction.values())

diction['age'] = 23  #updating existing key pair
print(diction['age']) 

diction['SKill'] = 'Python'  #adding new key pair

print(diction.items()) #getting items as list of a dictionary

del diction['Address'] #to delete any key value pair
print(diction)

removed_item = diction.pop('age') #to delete any key value pair and get the value
print(removed_item)

squareNum = {x:x**2 for x in range(5)} #Dictionary Comprehensive uses
print(squareNum)
```

```python
#Unit - 3 : NumPy (Numeric Python)

#ndArray : n-Dimensional Array

import numpy as np

onedarray = np.array([10,20,30])
twodarray = np.array([[10,20,30],[40,50,60]])
print(onedarray)

print(twodarray)
print("Type: ", type(twodarray)) #ndarray
print("Shape: ", twodarray.shape) #shape of array
print("Dimension: ", twodarray.ndim) #dimension of array
print("Size: ", twodarray.size) #size of array
print("Data type: ", twodarray.dtype) #data type of array elements
print("Length: ", len(twodarray)) #length of array (number of rows)
print("Element size: ", twodarray.itemsize) #size of each element in bytes

print("index 1: ", onedarray[1]) #indexing
print(twodarray[0,2]) #indexing

#slicing
print(onedarray[1:2]) #prints 2nd element
print(twodarray[:2]) #prints both rows
print(twodarray[1:]) #prints 2nd row
print(twodarray[:,1:]) #prints all rows and 2nd and 3rd column
```

```python
#npiter

import numpy as np

for item in np.nditer(twodarray):
    print(item)

#ndenumerate of elements in array

for idx, item in np.ndenumerate(twodarray):
    print(idx, item)
```

```python
#conditional array

import numpy as np

cond_array = np.array([10,15,20,25,30,35,40])

print("Original Array: ", cond_array)
print(cond_array[...,5]) #specific index
print(cond_array[[0,2,4]]) #specific index
print(cond_array[cond_array%2==0]) #even numbers
print(cond_array[cond_array>20]) #numbers greater than 20
```

```python
#Operation of NumPy

import numpy as np
array1 = np.array([10,20,30,40])
array2 = np.array([1,2,3,4])
print("Addition: ", array1 + array2)
print("Subtraction: ", array1 - array2)
print("Multiplication: ", array1 * array2)
print("Division: ", array1 / array2)
print("Exponentiation: ", array1 ** array2)
print("Square Root: ", np.sqrt(array1))
print("Absolute values: ", np.abs(array1))
print("Round values: ", np.round(array1/2))
print("Sin values: ", np.sin(array1))
print("Cos values: ", np.cos(array1))
print("Log values: ", np.log(array1))
print("Max value: ", np.max(array1))
print("Min value: ", np.min(array1))
print("Sum of all elements: ", np.sum(array1))
print("Mean value: ", np.mean(array1))
print("Standard Deviation: ", np.std(array1))
print("Variance: ", np.var(array1)) #variance of array elements
print("Dot Product: ", np.dot(array1, array2))
```

```python
#Unit - 4 : Pandas (Python Data Analysis Library)

import pandas as pd
data = {'Name': ['Rupak', 'Rishabh', 'Vivek', 'Prashant'],
        'Age': [22, 23, 21, 24],
        'City': ['Bhilai', 'Raipur', 'Durg', 'Bilaspur']}

df = pd.DataFrame(data)
print(df)
print(df.head(3)) #first 5 rows
print(df.tail(2)) #last 2 rows
print(df.columns) #column names
print(df.describe()) #statistical summary
print(df.info()) #dataframe info
print("Name: ", df['Name']) #accessing a column
print(df.iloc[1]) #accessing a row by index
print(df.loc[0, 'City']) #accessing a specific value

df['Age'] = df['Age'] + 1 #updating a column
print(df)
df['Country'] = 'India' #adding a new column
print(df)
df = df.drop('City', axis=1) #deleting a column
print(df)
df = df[df['Age'] > 22] #filtering rows
print(df)
df = df.sort_values(by='Name') #sorting dataframe
print(df)
print(df.isnull().sum()) #checking for null values
df.to_csv('output.csv', index=False) #exporting to CSV file
print("Data exported to output.csv")
```

```python
#Pandas series
import pandas as pd
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
print(series[2]) #accessing element
print(series.describe()) #statistical summary
print(series.mean()) #mean value
print(series.std()) #standard deviation
print(series.sum()) #sum of elements
print(series.max()) #maximum value
print(series.min()) #minimum value
print(series.value_counts()) #count of unique values
print(series.head(3)) #first 3 elements
print(series.tail(2)) #last 2 elements
series.to_csv('series_output.csv', index=False) #exporting series to CSV file
print("Series exported to series_output.csv")
```

```python
#Unit - 5 : Data Visualization using Matplotlib

import matplotlib.pyplot as plt

x = np.linspace(0.2, 2*np.pi, 10)
plt.stem(x, np.cos(x), linefmt='g-', markerfmt='D', basefmt='r-') #stem plot with green lines, diamond markers, and red baseline
plt.title("Stem Plot of Cosine Function")
plt.show()

x = [1, 2, 3, 4, 5]
z = [3, 5, 2, 1, 0]
y = [10, 20, 25, 30, 40]

plt.plot(x, z, color="b", linestyle='-', marker='.', label='Line Plot') #line plot
plt.plot(x, y, color="r", linestyle='--', marker='o', label='Dashed Line Plot') #line plot
plt.bar(x, y, color="g", label='Bar Plot') #bar plot
plt.title("Line and Bar Plot Example")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.annotate("Sample Annotation", xy=(3,25), xytext=(3,35), arrowprops=dict(facecolor='black', shrink=0.02)) #annotation with arrow
plt.grid()
plt.legend()
plt.show()

plt.pie(y, labels=x, autopct='%1.1f%%', startangle=100, explode=[0.0, 0.0, 0.1, 0.2, 0.0], shadow=True, radius=0.7) #pie chart #autopct to show percentage, startangle to rotate start position
plt.title("Pie Chart Example")
plt.legend()
plt.show()

plt.hist(y, bins=4, color='c', edgecolor='black') #histogram
plt.title("Histogram Example")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.legend()
plt.show()

plt.subplot(2, 1, 1) #2 rows, 1 column, 1st plot
plt.plot(x, y, color='m', label='First Plot')
plt.title("Subplot Example")
plt.legend()
plt.subplot(2, 1, 2) #2 rows, 1 column, 2nd plot
plt.bar(x, z, color='y', label='Second Plot')
plt.legend()
plt.show()

plt.stem(x, y, label='Midpoint Step') #stem plot
plt.step(x, y, where='mid', label='Midpoint Step') #step plot
plt.title("Step and stem Plot Example")
plt.legend()
plt.show()
```
